# **PROJECT** : Car Price Prediction

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedShuffleSplit,cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import joblib
from sklearn.metrics import r2_score
```

## Reading the Data

```python
df=pd.read_csv("cardekho_dataset.csv")
```

```python
df.head()
```

## Analysing the Data

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop("car_name",axis=1)
```

```python
df.describe()
```

```python
df.hist(bins=50,figsize=(12,8))
```

## Train Test Split

```python
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_indices,test_indices in split.split(df,df["fuel_type"]):
    train_set=df.loc[train_indices]
    test_set=df.iloc[test_indices]
    
```

## seprating the features and label

```python
x_train=train_set.drop("selling_price",axis=1)
y_train=train_set["selling_price"]
```

## Preprocessing


```python
# Numerical Columns
num_col=["vehicle_age","km_driven","mileage","engine","max_power","seats"]
```

```python
# Categorial Columns
cat_col=["brand","model","seller_type","fuel_type","transmission_type"]
```

```python
# Numerical Pipeline
num_pipe=Pipeline([
    ("imput",SimpleImputer(strategy="median")),
    ("standardscaler",StandardScaler())
])
```

```python
# Categorial Pipeline
cat_pipe=Pipeline([
    ("onehot",OneHotEncoder(handle_unknown="ignore"))
])
```

```python
# Combining the Pipeline
full_pipeline=ColumnTransformer([
    ("num",num_pipe,num_col),
    ("cat",cat_pipe,cat_col)
])
```

## Combining our Model

```python
# importing all Models

from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
# Random Forest Regrssor
ran_reg=Pipeline([
    ("Preprocess",full_pipeline),
    ("model",RandomForestRegressor())
])
```

```python
# Gradient Boosting Regressor
grad_reg=Pipeline([
    ("Preprocess",full_pipeline),
    ("model",GradientBoostingRegressor())
])
```

```python
# Linear Regression
lin_reg=Pipeline([
    ("Preprocess",full_pipeline),
    ("model",LinearRegression())
])
```

```python
# Decision Tree Regressor
dec_reg=Pipeline([
    ("Preprocess",full_pipeline),
    ("model",DecisionTreeRegressor())
])
```

```python
svm_reg=Pipeline([
    ("Preprocess",full_pipeline),
    ("model",SVR())
])
```

## Testing Our Model by Cross Val Score

```python

ran_rmses=-cross_val_score(ran_reg,x_train,y_train,scoring="neg_root_mean_squared_error",cv=10)
pd.Series(ran_rmses).describe()
```

```python
grad_rmses=-cross_val_score(grad_reg,x_train,y_train,scoring="neg_root_mean_squared_error",cv=10)
pd.Series(grad_rmses).describe()
```

```python
lin_rmses=-cross_val_score(lin_reg,x_train,y_train,scoring="neg_root_mean_squared_error",cv=10)
pd.Series(lin_rmses).describe()
```

```python
dec_rmses=-cross_val_score(dec_reg,x_train,y_train,scoring="neg_root_mean_squared_error",cv=10)
pd.Series(dec_rmses).describe()
```

```python
svm_rmses=-cross_val_score(svm_reg,x_train,y_train,scoring="neg_root_mean_squared_error",cv=10)
pd.Series(svm_rmses).describe()
```

## Saving the best Model

```python
# As RandomForest Shows the best result saving that

ran_reg.fit(x_train,y_train)
```

```python
joblib.dump(ran_reg,"model.pkl")
```

## Now testing Our Test Data for infrences

```python
x_test=test_set.drop("selling_price",axis=1)
y_test=test_set["selling_price"]
```

```python
y_predict=ran_reg.predict(x_test)
```

```python
print(f"r2 score:{r2_score(y_test,y_predict)}")
```
